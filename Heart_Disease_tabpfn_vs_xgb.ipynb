```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
```

```python
import mlcroissant as mlc

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/cherngs/heart-disease-cleveland-uci/croissant/download')

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
record_set_df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
record_set_df.head()
```

```python
print(record_set_df.columns.tolist())

df = record_set_df
```

```python
# Fix column names
df.columns = df.columns.str.replace(r'^.*?/', '', regex=True)
print(df.columns.tolist())
```

```python
# Separate features and target
X = df.drop(columns=['condition'])
y = df['condition']
```

```python
# Column classification
categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Target
target_col = 'condition'
```

```python
# Split
X = df.drop(columns=[target_col])
y = df[target_col]

# Encode categorical features
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
import numpy as np

encoder = OrdinalEncoder()
X_cat = encoder.fit_transform(X[categorical_cols])

# Scale numerical features
scaler = StandardScaler()
X_num = scaler.fit_transform(X[numerical_cols])

# Combine
X_processed = np.hstack([X_num, X_cat])
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
```

```python
from tabpfn import TabPFNClassifier
from sklearn.metrics import accuracy_score, classification_report


clf_tabpfn = TabPFNClassifier(device='cpu')
clf_tabpfn.fit(X_train, y_train)
y_pred_tabpfn = clf_tabpfn.predict(X_test)

print("TabPFN Accuracy:", accuracy_score(y_test, y_pred_tabpfn))
print(classification_report(y_test, y_pred_tabpfn))
```

```python
from xgboost import XGBClassifier

clf_xgb = XGBClassifier(eval_metric='logloss', random_state=42)
clf_xgb.fit(X_train, y_train)
y_pred_xgb = clf_xgb.predict(X_test)

print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt

# List of features in the order they were fed into the model
features = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
    'exang', 'oldpeak', 'slope', 'ca', 'thal'
]

# Human-readable mapping
feature_mapping = {
    'age': 'Age (years)',
    'sex': 'Sex (1 = male, 0 = female)',
    'cp': 'Chest pain type (03)',
    'trestbps': 'Resting blood pressure (mm Hg)',
    'chol': 'Serum cholesterol (mg/dl)',
    'fbs': 'Fasting blood sugar > 120 mg/dl (1 = true, 0 = false)',
    'restecg': 'Resting ECG results (02)',
    'thalach': 'Max heart rate achieved',
    'exang': 'Exercise-induced angina (1 = yes, 0 = no)',
    'oldpeak': 'ST depression induced by exercise',
    'slope': 'Slope of peak exercise ST segment (02)',
    'ca': 'Number of major vessels (04) colored by fluoroscopy',
    'thal': 'Thalassemia (1 = normal, 6 = fixed defect, 7 = reversible defect)'
}

# Plot importance
ax = plot_importance(clf_xgb, max_num_features=10)
ax.set_title("Top 10 Features - XGBoost")

# Replace f# labels with readable names
yticklabels = ax.get_yticklabels()
for label in yticklabels:
    feature_idx = int(label.get_text()[1:])  # remove "f" and convert to int
    feature_name = features[feature_idx]
    label.set_text(feature_mapping[feature_name])

ax.set_yticklabels(yticklabels)
plt.show()
```
